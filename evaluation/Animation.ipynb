```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

t = np.linspace(0,2*np.pi)
x = np.sin(t)
x1 = np.cos(t)
x2 = np.tan(t)

y = np.array([x,x1,x2])
y.shape
```

```python
fig, ax = plt.subplots()
l, = ax.plot([0,2*np.pi],[-1,1])

def animate(i):
    
    l.set_data(t, y[i,:])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))

plt.show()
```
